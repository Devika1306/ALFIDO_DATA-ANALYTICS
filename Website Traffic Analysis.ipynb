```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
file_path = r'C:\Users\hp\Downloads\website_wata (1).csv'
df = pd.read_csv(file_path)
```

```python
print("First few rows of the dataset:")
print(df.head())
```

```python
print("\nBasic Information:")
print(df.info())
```

```python
# Summary statistics of the dataset
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Distribution of numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
print("\nDistribution of Numerical Features:")
df[numerical_features].hist(bins=15, figsize=(15, 10))
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
# Distribution of categorical features
categorical_features = df.select_dtypes(include=['object']).columns
print("\nDistribution of Categorical Features:")
for feature in categorical_features:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=feature, data=df)
    plt.title(f"Distribution of {feature}")
    plt.show()
```

```python
# Correlation matrix
correlation_matrix = df[numerical_features].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Pairplot of the dataset
print("\nPairplot of the Dataset:")
sns.pairplot(df[numerical_features])
plt.show()
```

```python
# Boxplots to check for outliers in numerical features
print("\nBoxplots of Numerical Features:")
for feature in numerical_features:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=feature, data=df)
    plt.title(f"Boxplot of {feature}")
    plt.show()
```

```python
# Distribution of 'Conversion Rate' based on 'Traffic Source'
print("\nConversion Rate based on Traffic Source:")
plt.figure(figsize=(10, 5))
sns.boxplot(x='Traffic Source', y='Conversion Rate', data=df)
plt.title("Conversion Rate based on Traffic Source")
plt.show()
```
